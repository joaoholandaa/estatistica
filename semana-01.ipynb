## 1. Simulador do lucro médio da promoção da coxinha

```python
'''
Exercício 1: Seja uma probabilidade p (amostral) de um cliente não pagar a conta, e a loja ter um prejuízo de R$ 33,
e que a promoção aumenta o lucro de Lm, calcular o lucro médio.
'''

# Leitura dos dados (serão do tipo string)
La = input('Digite o lucro atual: ')
Lm = input('Digite o lucro adicional: ')
p = input('Prob. de perda: ')

# Convertendo os dados para números com casas decimais (float)
La = float(La)
Lm = float(Lm)
p = float(p)

# Usando os valores na fórmula do lucro novo
Ln = La - 33*p + 33*(1-p) + Lm

# Imprimindo o resultado contido na variável Ln
print('Seu lucro novo será de:', Ln)
```
