# 1. Construindo um gráfico de barras para a frequência <br>
A sequência de comandos a seguir ilustra como criar um gráfico de barras.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6]
y = [50, 100, 200, 200, 100, 50]

plt.bar(x,y,color='red')
plt.grid()
```

# 2. Transformando gráfico de frequência em probabilidade <br>
Para transformar os valores de frequência em frequência relativa, que pode ser considerada uma aproximação para probabilidade, é necessário:

1. Dividir a frequência de cada face pelo total de experimentos realizados;

2. Criar o gráfico da frequência relativa de cada face.


```python
total = sum(y)
p = []
for yi in y:
    p.append(yi/total)
p
```

```python
plt.bar(x,p,color='magenta')
plt.grid()
```

# 3. Simulador de um dado de seis faces <br>
É possível criar um simulador de um dado de seis face através de um comando que simula o sorteio de um valor aleatório inteiro em um intervalo [a,b] através do comando random.randint(a,b). Os vários valores gerados podem ser armazenados em uma lista inicialmente vazia através do comando append.

```python
import random

val = []
for i in range(0,31):
    num = random.randint(1,6)
    val.append(num)

print(val)
```

Para criar o gráfico de frequência de ocorrência de cada face pode ser utilizado o comando np.histogram.

```python
import matplotlib.pyplot as plt
import numpy as np

freq, bin_edges = np.histogram(val,bins=range(1,8))
print(freq)
plt.bar(x,freq)
plt.grid()
```

# 4. Agrupando valores em classes para construir um histograma <br>

É possível agrupar os valores anteriormente gerados em três classes que compreendem três faixas de valores:

1. Classe A: [1, 2]

2. Classe B: [3, 4]

3. Classe C: [5, 6]

Para tanto, basta voltar a utilizar o comando hist e o parâmetro bins deve compreender as faixas de valores das classes através dos seguintes valores: [1, 3, 5, 7 ]. É possível ainda contabilizar o número de valores em cada faixa com o comando np.histogram.


```python
import matplotlib.pyplot as plt
import numpy as np

freq, bin_edges = np.histogram(val, bins=[1,3,5,7])
print(freq)
plt.hist(val, density=True, bins=[1,3,5,7])
plt.grid()
```
